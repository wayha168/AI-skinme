```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the data (replace 'your_data_file.csv' with the name of your file)
df = pd.read_csv('data/Lumpy skin disease data.csv')

# 2. Execute the plot command
plt.figure(figsize=(12, 8))
sns.scatterplot(x='x', y='y', hue='lumpy', data=df, palette='viridis', s=20)
plt.title('Geographical Distribution of Skin Check Outcome (lumpy)')
plt.xlabel('Longitude (x)')
plt.ylabel('Latitude (y)')
plt.legend(title='Lumpy Status')
plt.savefig('assets/geographical_lumpy_distribution.png')
plt.show()
```

```python

# --- STEP 1: Load Data from CSV (The Fix) ---
try:
    # Use your actual file loading command
    df = pd.read_csv('data/Lumpy skin disease data.csv')
    
    # Crucial: Ensure the 'lumpy' column is treated as a numerical type (0 or 1)
    if 'lumpy' in df.columns:
        # Convert to integer, handling potential non-numeric entries (e.g., NaNs or strings)
        df['lumpy'] = pd.to_numeric(df['lumpy'], errors='coerce').fillna(0).astype(int)
    
    print("--- Real LSD Data Loaded Successfully ---")
    print(df.head())
    print("\n--- Data Types ---")
    print(df.info())

except FileNotFoundError:
    print("ERROR: CSV file not found at 'data/Lumpy skin disease data.csv'.")
    print("Please check the file path and name, or make sure the file is in the correct directory.")
    # Exit or handle the error gracefully if the file is essential
    raise # Re-raise the error to stop execution if the file is missing

# --- STEP 2: Filter the Data ---
# Define the regions you want to focus on (Asia, Europe, Americas)
target_regions = ['Asia', 'Europe', 'Americas'] 

# Create the filtered DataFrame
# This step is critical because your plotting command uses 'df_filtered'
if 'region' in df.columns:
    df_filtered = df[df['region'].isin(target_regions)].copy()
    print(f"\nFiltered Data Points (Asia, Europe, Americas): {len(df_filtered)}")
else:
    print("ERROR: 'region' column not found in the loaded CSV. Cannot filter.")
    df_filtered = df.copy() # Proceed without filtering if column is missing


# --- STEP 3: Execute the plot command on the FILTERED data ---

# Check if essential columns for plotting exist
if all(col in df_filtered.columns for col in ['x', 'y', 'lumpy']):
    plt.figure(figsize=(15, 10))
    sns.scatterplot(
        x='x', 
        y='y', 
        hue='lumpy', # Your outcome variable
        data=df_filtered, 
        palette='viridis', # Good for continuous or binary data visualization
        s=30,
        alpha=0.7 
    )
    plt.title('Geographical Distribution of Lumpy Status (Filtered: Europe, Asia, Americas)')
    plt.xlabel('Longitude (x)')
    plt.ylabel('Latitude (y)')
    plt.legend(title='Lumpy Status')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.savefig('assets/distrubution_lumpy_status.png')
    plt.show()
else:
    print("\nERROR: Cannot plot. Missing one or more essential columns ('x', 'y', or 'lumpy') in the DataFrame.")
```
